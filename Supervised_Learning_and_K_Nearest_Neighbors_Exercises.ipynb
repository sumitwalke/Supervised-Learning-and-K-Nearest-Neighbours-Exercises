```python
from sklearn.preprocessing import LabelBinarizer
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('/content/drive/MyDrive/TY MLL/Orange_Telecom_Churn_Data.csv')
```

```python
df.head(10)
```

```python
df.columns
```

```python
df = df.drop(columns = ['state', 'area_code', 'phone_number'])
```

```python
lb = LabelBinarizer()

categorical_columns = ['intl_plan', 'voice_mail_plan', 'churned']

for col in categorical_columns:
    df[col] = lb.fit_transform(df[col])

print("\nDataset after encoding categorical features:")
df.head()
```

```python
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(df.drop(columns=['churned']))
scaled_df = pd.DataFrame(scaled_features, columns=df.columns[:-1])
scaled_df['churned'] = df['churned']
print("\nDataset after scaling features:")
scaled_df.head()
```

```python
X = scaled_df.drop(columns = ['churned'])
y = scaled_df['churned']
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)

y_pred = knn.predict(X)

predicted_df = pd.DataFrame({'Actual Churned': y, 'Predicted Churned': y_pred})
print("\nPredicted outcomes on the same data:")
predicted_df.head()
```

```python
correct_predictions = np.sum(y == y_pred)
total_predictions = len(y)
accuracy = correct_predictions / total_predictions
print("Accuracy of the model", accuracy*100, '%')
```

```python
knn_distance = KNeighborsClassifier(n_neighbors = 3, weights = 'distance')
knn_distance.fit(X, y)
y_pred_distance = knn_distance.predict(X)
```

```python
correct_predictions = np.sum(y == y_pred_distance)
total_predictions = len(y)
accuracy = correct_predictions / total_predictions
print("Accuracy of the model", accuracy*100, '%')
```
